Se creo la columna 'release_year'

```python
df['release_year']=df['release_date'].dt.year
```

Fue creada una nueva columna: 'return' , esta columna tendra los valores que surjan de la relacion entre  revenue/budget , pero previamente se chequeo si esa operacion era posible

```python
df.info()
```

Cambio del tipo de dato en la columna 'budget'

```python
df['budget']=pd.to_numeric(df['budget'],downcast='float', errors='coerce')
```

```python
df['return']=df['revenue'].divide(df['budget']).fillna(0)
```

```python
df
```

Los valores 'inf' fueron reemplazados por Nan y luego por 0

```python
df['return'].replace([np.inf,-np.inf], np.nan, inplace= True)
```

```python
df['return'].fillna(0,inplace=True)
```

Existian dos columnas con diccionarios y tres columnas con listas , a su vez las mismas podian contener en algunos casos, listas dentro, se procedio a desanidar las mismas con el modulo 'ast'

```python
df.head()
```

```python
import ast
```

```python
def convert(obj):
    if isinstance(obj,str) and '{'in obj:
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name'])
        return L
df["genres"]=df["genres"].apply(convert)
```

```python
df
```

```python
df['production_companies']=df['production_companies'].apply(convert)
df['production_countries']=df['production_countries'].apply(convert)
df['spoken_languages']=df['spoken_languages'].apply(convert)
```

```python
df
```

```python
def convert1(obj):
    if isinstance(obj, str) and '{' in obj:
        dic = ast.literal_eval(obj)
        return dic['name']
df['belongs_to_collection']=df['belongs_to_collection'].apply(convert1)
```
